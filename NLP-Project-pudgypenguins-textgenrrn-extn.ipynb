```python
import requests

url = "https://api.opensea.io/api/v1/collection/pudgypenguins"
response = requests.get(url)
print(response.text)
```

```python
import textgenrnn 
textgen = textgenrnn.textgenrnn()
#https://github.com/minimaxir/textgenrnn
```

```python
textgen.reset()

### as before I have manually taken the 'traits' from the api call and collated them onto a csv file

with open("data/pudgypenguinstraits.csv") as f:
    content = f.readlines()
    
content
```

```python
textgen.reset()
textgen.train_on_texts(content, num_epochs=10,  gen_epochs=2)
```

```python
textgen.generate(5)
```

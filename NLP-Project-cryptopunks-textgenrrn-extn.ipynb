```python
import textgenrnn 
textgen = textgenrnn.textgenrnn()
#https://github.com/minimaxir/textgenrnn
```

```python
textgen.reset() # I need to call this everytime I want to retrain/reset

### as before I have manually taken the 'traits' from the api call and collated them onto a csv file

with open("data/cryptopunkstraits-accessories-column.csv") as f:
    content = f.readlines()
    
content
```

```python
textgen.reset()
textgen.train_on_texts(content, num_epochs=10,  gen_epochs=2)
```

```python
textgen.generate(25)
#some of which that were generated here are not in the line up!
## I've noticed the closer the loss gets to 0.5, there is too much of an identical likeness to the input.

#It is taking some of the traits and reorganising the words, which in essense does satify the aim I am getting at. 
# for example 'mohawk blonde', and 'black shades' is not in the original list. 
```
